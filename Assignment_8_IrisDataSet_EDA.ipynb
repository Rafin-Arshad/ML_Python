### PDF(Probability Density Function) and CDF(Cumulative Density Function)

```python
pdf = count/sum(count)
```

```python
pdf ## setosa's pdf
```

```python
# virginica count and bin_edge
```

```python
count_virginica, bin_edge_virginica = np.histogram(df_virginica['PetalLengthCm'],density=True,bins=10)
```

```python
print(count_virginica)
print(bin_edge_virginica)
```

```python
df_virginica.describe() 
```

```python
#pdf of virginica
pdf_virginica = count_virginica/sum(count_virginica)
pdf_virginica
```

##### pdf plot of setosa's bin_edge

```python
plt.plot(bin_edge[1:],pdf)
```

##### pdf of virginica's bin_edge

```python
plt.plot(bin_edge_virginica[1:],pdf_virginica)
```

#### CDF of setosa

```python
cdf_setosa = np.cumsum(pdf)
plt.plot(bin_edge[1:],cdf_setosa)
```

#### cdf of virginica

```python
cdf_virginica = np.cumsum(pdf_virginica)
plt.plot(bin_edge_virginica[1:],cdf_virginica)
```

```python
# Plots of CDF of petal_length for various types of flowers.

# Misclassification error if you use petal_length only.
import numpy as np
import seaborn as sns
counts, bin_edges = np.histogram(df_setosa['PetalLengthCm'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


# virginica
counts, bin_edges = np.histogram(df_virginica['PetalLengthCm'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


#versicolor
counts, bin_edges = np.histogram(df_versicolor['PetalLengthCm'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


plt.show();
```

```python
# Plots of CDF of petal_length for various types of flowers.

# Misclassification error if you use petal_width only.
import numpy as np
import seaborn as sns
counts, bin_edges = np.histogram(df_setosa['PetalWidthCm'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


# virginica
counts, bin_edges = np.histogram(df_virginica['PetalWidthCm'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


#versicolor
counts, bin_edges = np.histogram(df_versicolor['PetalWidthCm'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


plt.show();
```
